```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline

from pylab import rcParams
rcParams['figure.figsize'] = 10, 10
```

```python
train_log = pd.read_csv("./mnist.log.train")
test_log = pd.read_csv("./mnist.log.test")
plt.figure(figsize=(8, 8))
plt.xlabel('iteration')
plt.ylabel('accuracy')

#plt.plot(train_log["NumIters"], train_log["accuracy"], color = 'red', label='train')
plt.plot(test_log["NumIters"], test_log["accuracy"], color = 'blue', label='test')
plt.legend(loc = 'lower right')
plt.show()
```

```python
train_log = pd.read_csv("./mnist.log.train")
test_log = pd.read_csv("./mnist.log.test")
plt.figure(figsize=(8, 8))
plt.xlabel('iteration')
plt.ylabel('loss')

plt.plot(train_log["NumIters"], train_log["loss"], color = 'red', label='train')
plt.plot(test_log["NumIters"], test_log["loss"], color = 'blue', label='test')
plt.legend(loc = 'lower right')
plt.show()
```
